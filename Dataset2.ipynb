### Cosine similarity score

```python
from sklearn.feature_extraction.text import TfidfVectorizer
netflix_data = netflix.copy()
tfidf = TfidfVectorizer(stop_words='english')
netflix_data['description'] = netflix_data['description'].fillna('')
tfidf_matrix=tfidf.fit_transform(netflix_data['description'])
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
cosine_sim
```

```python
# Create a function to check the most popular cast, directors and genres
def most_popular(df, col):
    df[col].value_counts().head(10).plot(kind='barh', color=['maroon', 'pink'])
    plt.title('Top 10 most popular ' + col)
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
    
# Apply the function to check the most popular cast, directors and genres for movies and TV shows
most_popular(df_movie, 'cast')
most_popular(df_movie, 'director')
most_popular(df_movie, 'listed_in')
most_popular(df_tvshow, 'cast')
most_popular(df_tvshow, 'director')
most_popular(df_tvshow, 'listed_in')
```

### Why Data Analysis is important?
###### Netflix has approximately 213.6 million paid subscribers worldwide as of the 3rd quarter of 2021. Netflix is the #1 Over-the-Top (OTT) platform since its launch in 2007.  Again, in the 3rd quarter of 2021, Netflix generated total revenue of nearly 7.5 billion U.S. dollars. This is all because of data analytics.

###### By Data Analysis Netflix can find out what content is available in different countries, can identify similar content by text-based features, analyse Actors/Directors and find interesting insights, can compare total counts of TV Shows and Movies in recent years to find out what is trending etc.

###### We plotted bar chart of number of movies and number of TV Shows in different countries, this helps to analyse what kind of content people are interested in and what needs to be promoted and released more.  Figuring out top 10 genres in each countries is important for licensing costs. Data of titles are collected and analysed which movie or TV Show title had biggest viewership relative to the licensing cost, if any titles are not watched enough relative to their cost then that show is not renewed. Most of us have a preference towards a few genres. While some may like adventure, action, crime, thrillers etc., others might prefer romance, comedy or family movies. While we are browsing for titles, most of us would probably anchor our search by genre. So, I picked genre as an important variable to figure out how Netflix adds its movies and tv shows. Every title is usually classified in about 23 genres. 

######  If we are watching a series like say abc, Netflix is able to see (on a large scale) the completion rate (for lack of a better term) of users. For example, the people at Netflix could ask themselves How many users who started abc (from season 1) finished it to the end of season let's say  3? Then they get an answer. Lets say its 70%.

###### Then they ask Where was the common cut off point for users? What did the other 30% of users do? How big of a time gap was there between when consumers watched one episode and when they watched the next? We need to get a good idea of the overall engagement of this show. They then gather this data and see user trends to understand engagement at a deep level. If Netflix saw that 70% of users watched all seasons available of a cancelled show, that may provoke some interest in restarting abc. They know theres a good chance users will watch the new season. 

###### Keeping track of viewership with respect to duration is important, to understand  what timeframe the viewers prefer. 
###### By plotting number of movies and TV Shows added every year, Netflix can figure out viewership, favourite director or genre in any particular year.
